```python
# we need two dimensional dataset so we add the two square brackest 
x= df[["CGPA"]]
# to see the demnensions

```

```python
y=df["Package (LPA)"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="CGPA", y="Package (LPA)", data=df)
plt.show()
```

# Model selection 

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=45)
# you can change teh random state value to increace the accuracy of the model 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# BEST FIT LINE 
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
# predit the package
lr.predict([[6.9]])
```

```python
# ACCURAY SCORE 
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# fit line 
sns.scatterplot(x="CGPA", y="Package (LPA)", data=df)
plt.plot(df["CGPA"],lr.predict(x),c="red")
plt.legend(["org","predict line"
           ])
plt.show()
```
